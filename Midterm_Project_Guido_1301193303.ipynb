### Soal 1 (15 poin)
Tampilkan visualisasi distribusi data untuk variabel "Glucose" dan "Insulin", lalu jelaskan hasil distribusinya.



#### Code

```python
#pemanggilan kolom glucose dan insulin
df1 = df[['Glucose','Insulin']]
df1.head(10)
```

```python
# Histogram kolom glucose dan insulin
col = df1.columns[:2]
plt.subplots(figsize = (30, 5))
length = len(col)

for i, j in itertools.zip_longest(col, range(length)):
    plt.subplot((length/2), 3, j + 1)
    plt.subplots_adjust(wspace = 0.1,hspace = 0.5)
    df1[i].hist(bins = 10)
    plt.ylabel('Jumlah Orang')
    plt.xlabel('Tingkat')
    plt.title(i)
plt.show()
```

### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
#Function trend analysis
def trend_analysis(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    else:
        return "Bear drop"
```

### Saham Nasdaq

```python
#Mengecek data null
Nasdaq_df.isna().sum()
```

```python
#Day-to-day percentage change(Daily returns)
Nasdaq_df["Daily_Percentage_Change"] = Nasdaq_df["Adj Close"].pct_change()*100
# drop null values
Nasdaq_df.dropna(inplace = True, axis = 0)
#Trend Analysis
Nasdaq_df["Trend"]= np.zeros(Nasdaq_df['Daily_Percentage_Change'].count())
Nasdaq_df["Trend"]= Nasdaq_df["Daily_Percentage_Change"].apply(lambda x:trend_analysis(x))
#Set Date/year 2019
Nasdaq_df['Date'] = pd.to_datetime(Nasdaq_df['Date'])
Nasdaq_df_2019 = Nasdaq_df[Nasdaq_df['Date'].dt.year == 2019]
Nasdaq_df_2019.head()
```

```python
#Visualisasi Trend Frequency dengan Pie-Chart
Nasdaq_df_pie_data = Nasdaq_df_2019.groupby('Trend')
pie_label = sorted([i for i in Nasdaq_df_2019.loc[:, 'Trend'].unique()])
plt.pie(Nasdaq_df_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham Nasdaq pada tahun 2019', y = 1.5)
plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
#Set Index
Nasdaq_df_2019.set_index('Date', inplace=True)
Nikkei_df_2019.set_index('Date', inplace=True)
HangSeng_df_2019.set_index('Date', inplace=True)
```

```python
# merge colomn Adj close values
combined_df = pd.concat([Nasdaq_df_2019["Adj Close"], Nikkei_df_2019["Adj Close"], HangSeng_df_2019["Adj Close"]], axis=1)
combined_df.columns = ["Nasdaq", "Nikkei", "HangSeng"]
combined_df.head()
```

```python
#Mengecek data null
combined_df.isna().sum()
```

```python
# drop null values
combined_df.dropna(inplace = True, axis = 0)

# Day-to-day percentage change(Daily returns)
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

#Visualisasi Volatilitas dengan line-chart
combined_vol = pct_chg_df[["Nasdaq", "Nikkei", "HangSeng"]].rolling(7).std()*np.sqrt(7)
combined_vol.plot(figsize = (15, 7), marker ='o')
plt.title('Volatilitas Saham Nasdaq, Nikkei, Hang Seng pada tahun 2019')
plt.xlabel('Tahun-Bulan')
plt.ylabel('Harga')
```

#### Penjelasan 
Pertama melakukan set index 'date' untuk ketiga dataset saham yang sudah di set tahun 2019, melakukan merge atau penggabungan ketiga data set terhadap colom adj close tersebut, melakukan pengecekkan data null dan dilakukan mendrop data null pada dataset combine, seterusnya mengitung untuk pengembalian harian harga saham dari dataset yang sudah di combine, selanjutnya mengitung rata-rata 7 hari (moving average) dari pengembalian harian kemudian menghitung standar deviasi (yang merupakan akar kuadrat dari varians) dan memplot nilainya menggunakan fungsi rolling() dan std().

Dari Visualisasi Volatilitas dengan line-chart, bisa diamati/dilihat dari ketiga saham tersebut memiliki harga relatif cenderung sama antara range harga 1-4 dan bisa dilihat juga pada trend analysis dari ketiga saham tersebut yaitu trend Slight or No Change. Terjadi sekitar antara 2019-08 dengan 2019-09 pada saham Nasdaq memilik harga tertinggi dan untuk harga saham terendah terjadi pada saham Nikkei yang terjadi sekitar antara 2019-10 dengan 2019-11.
